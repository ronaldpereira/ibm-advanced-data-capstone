```python
import pandas as pd
import matplotlib.pyplot as plt
from shap import TreeExplainer, summary_plot

from red_wine_quality_prediction import XGBoostRegressionModel
```

```python
model = XGBoostRegressionModel(model_pickle_path='model/xgb_reg.pickle')
```

```python
shap_explainer = TreeExplainer(model.model['xgb_reg'])
```

```python
wine_quality_df = pd.read_csv('data/winequality-red.csv')
```

```python
shap_values = shap_explainer.shap_values(wine_quality_df.drop('quality', axis=1))
```

```python
summary_plot(shap_values, wine_quality_df.drop('quality', axis=1), show=False)
plt.tight_layout()
plt.savefig('images/red_wine_quality_shap_values.jpg', dpi=100)
```

```python
summary_plot(shap_values, wine_quality_df.drop('quality', axis=1), plot_type='bar', show=False)
plt.tight_layout()
plt.savefig('images/red_wine_quality_feature_importance.jpg', dpi=100)
```
